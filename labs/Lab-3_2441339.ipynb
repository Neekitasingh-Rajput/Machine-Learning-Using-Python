```python
#f. Fill nulls in object/categorical columns with 'na'

df_cat_filled = df.copy()

cat_cols = df_cat_filled.select_dtypes(include="object").columns
df_cat_filled[cat_cols] = df_cat_filled[cat_cols].fillna("na")

df_cat_filled
```

```python
#g. Fill nulls in numeric columns with mean

df_mean_filled = df.copy()

num_cols = df_mean_filled.select_dtypes(include=["float64", "int64"]).columns
df_mean_filled[num_cols] = df_mean_filled[num_cols].fillna(df_mean_filled[num_cols].mean())

df_mean_filled
```

```python
#h. Fill nulls in numeric columns with median

df_median_filled = df.copy()

num_cols = df_median_filled.select_dtypes(include=["float64", "int64"]).columns
df_median_filled[num_cols] = df_median_filled[num_cols].fillna(df_median_filled[num_cols].median())

df_median_filled
```
