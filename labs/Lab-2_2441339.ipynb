```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create dataset
data = {
    'Car_ID': ['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10','C11','C12'],
    'Brand': ['Maruti','Hyundai','Honda','Tata','Toyota','Maruti','Hyundai','Honda','Tata','Toyota','Maruti','Hyundai'],
    'Model_Year': [2015,2018,2017,2020,2016,2019,2021,2018,2016,2022,2023,2015],
    'Mileage_kmpl': [18.5,19.0,17.2,20.1,16.5,21.3,22.0,18.0,15.5,19.8,23.5,14.9],
    'Price_Lakh': [4.5,6.2,7.0,5.8,8.5,6.0,7.5, None,4.8,10.2,7.1,3.9]
}

df_cars = pd.DataFrame(data)
```

```python
# 1. First 5 rows
print("FIRST 5 ROWS:")
print(df_cars.head())
```

```python
# 2. Number of rows and columns
print("\nSHAPE (rows, columns):")
print(df_cars.shape)
```

```python
# 3. Summary statistics
print("\nSUMMARY STATISTICS for Mileage_kmpl & Price_Lakh:")
print(df_cars[['Mileage_kmpl','Price_Lakh']].describe())
```

```python
# 4. Maruti cars with Mileage > 20
print("\nMaruti with Mileage > 20:")
print(df_cars[(df_cars.Brand=='Maruti') & (df_cars.Mileage_kmpl>20)])
```

```python
# 5. Sort by Price in descending and show top 3
print("\nTop 3 cars by price:")
print(df_cars.sort_values(by='Price_Lakh', ascending=False).head(3))
```

```python
# 6. Fix missing value
print("\nRows with missing values:")
print(df_cars[df_cars.isna().any(axis=1)])

mean_price = df_cars['Price_Lakh'].mean()
df_cars['Price_Lakh'].fillna(mean_price, inplace=True)
```

```python
# 7. New column price per km
df_cars['Price_per_km'] = df_cars['Price_Lakh'] / df_cars['Mileage_kmpl']
print("\nFIRST 5 rows after adding Price_per_km:")
print(df_cars.head())
```

```python
# 8. Scatter plot
plt.figure(figsize=(6,5))
plt.scatter(df_cars['Mileage_kmpl'], df_cars['Price_Lakh'])
plt.xlabel("Mileage (kmpl)")
plt.ylabel("Price (Lakh)")
plt.title("Mileage vs Price")
plt.show()
```

```python
# 9. Histogram
plt.figure(figsize=(6,5))
plt.hist(df_cars['Mileage_kmpl'])
plt.xlabel("Mileage (kmpl)")
plt.ylabel("Frequency")
plt.title("Mileage Distribution")
plt.show()
```
