```python
#A. 
#Import the required library and load Students.csv into a Dataframe named df_students.
#Display the first 5 rows of the DataFrame.

import pandas as pd #necessary library

data = pd.read_csv(r"C:\Users\bca.lab21\Desktop\students.csv") #path of the file
df_students = pd.DataFrame(data)

df_students.head()
```

<h5>
    Obs: pandas library is the library that is used for Data Manipulation. df.head() gives us the view of first 5 rows in our Data.
</h5>

```python
#B.
#Write code to display the number of rows and columns in df_students.

print("(Rows, Columns) = ",df_students.shape)
```

<h5>
Obs: No. of Rows and Columns can be obtained from the shape of the dataset.
</h5>

```python
#C. 
#Obtain the summary statistics for the columns Internals and Finals. Write the code.

print("\nSummary statistics for the columns Internals and Finals:\n")
df_students[["Internals","Finals"]].describe()
```

<h5>
    Obs: Summary statistics of any numeric column is obtained using describe() function.
</h5>

```python
#D.
#Select and Display all rows where Attendance_Percent > 85.

df_students[df_students["Attendance_Percent"] > 85]
```

<h5>
    Obs: Filtering helps us view a required data with a specified condition.
</h5>

```python
#E.
#In the dataset, Attendance_Percent has a missing Value.

#(i). Write code to identify rows where Attendance_Percent is missing.

print("\nRow with missing Attendance Percent:")
df_students[df_students["Attendance_Percent"].isnull() == True]
```

<h5>
    Obs: isnull checks for us if there is any missing data in our dataset.
</h5>

```python
#(ii). Fill the missing Attendance_Percent with the mean attendance of the column.

mean_att = df_students["Attendance_Percent"].mean()
print("\nMean attendance:", mean_att)
df_students = df_students.fillna(mean_att)
df_students
```

<h5>
    Obs: mean can be calculated and then be imputed for numeric columns at the missing places.
</h5>

```python
#(iii). Create a new column Total_Marks = Internals + Finals and display the first 5 rows of the updated DataFrame.

df_students["Total Marks"] = df_students["Internals"] + df_students["Finals"]
df_students.head()
```

<h5>
    Obs: New columns with calculated columns help us infer more about the dataset.
