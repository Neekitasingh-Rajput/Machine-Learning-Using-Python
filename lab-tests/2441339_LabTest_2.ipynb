```python
#import necessasry libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#load the dataset

data = pd.read_csv("salaries.csv")
df = pd.DataFrame(data)
df
```

**Q2) Data Preprocessing**

```python
#1. Separate features X and target y (salary_more_then_100k)

X = df.drop('salary_more_then_100k',axis=1)
y = pd.DataFrame(df['salary_more_then_100k'])
```

```python
X
```

```python
y
```

```python
#2. Split data into train and test sets (recommended: 80/20 or 75/25) with

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
#3. Encode categorical variables

categorical_cols = df.select_dtypes(include = ['object']).columns
categorical_cols
```

```python
le = LabelEncoder()
df['degree'] = le.fit_transform(df['degree']) #as degree is an ordinal column
df = pd.get_dummies(df) #for other categorical columns
```

```python
df
```

```python
#4. Print the shape of X_train and X_test after preprocessing (number of rows and columns)
```

```python
print("Training dataset:", X_train.shape)
print("Test dataset:", X_test.shape)
```

**Q3) Model Building (Decision Tree)**

```python
dt = DecisionTreeClassifier(criterion = "entropy")
dt.fit(X_train,y_train)
```

**Q4) Model Evaluation**

```python
dt_pred = dt.predict(X_test)
dt_acc = accuracy_score(y_test, dt_pred)

print("Model Accuracy:", dt_acc)
```

```python
#Confusion matrix

cm = confusion_matrix(y_test, dt_pred)
print(cm)
```

```python
#classification_report

cr = classification_report(y_test, dt_pred)
print(cr)
```
